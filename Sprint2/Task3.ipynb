```
public interface IRentable
{
    void Rent();
}

public interface IStreamable
{
    void Stream();
}

public abstract class Movie
{
    public string Title { get; set; }
    public string Director { get; set; }
    public int ReleaseYear { get; set; }
    public int Rating { get; private set; }

    public Movie(string title, string director, int releaseYear)
    {
        Title = title;
        Director = director;
        ReleaseYear = releaseYear;
    }

    public abstract void Watch();
    
    public virtual void GetInfo()
    {
        Console.WriteLine($"{Title}, {Director}, {ReleaseYear}");
    }

    public virtual void Rate(int rating)
    {
        this.Rating = rating;
        Console.WriteLine($"Фильм оценен на {rating}/10");
    }
}

class Documentary : Movie, IStreamable
{
    public string Theme { get; set; }

     public Documentary(string title, string director, int releaseYear, string theme)
        : base(title, director, releaseYear)
    {
        Theme = theme;
    }

    public override void Watch()
    {
        Console.WriteLine("Вы смотрите документальный фильм");
    }

    public override void GetInfo()
    {
        Console.WriteLine($"{Title}, {Director}, {ReleaseYear}, Тема: {Theme}");
    }

    public void Stream()
    {
        Console.WriteLine($"Документальный фильм \"{Title}\" доступен для потоковой передачи.");
    }
}

class FeatureFilm : Movie, IRentable
{
    public string Genre { get; set; }

    public FeatureFilm(string title, string director, int releaseYear, string genre)
        : base(title, director, releaseYear)
    {
        Genre = genre;
    }

    public override void Watch()
    {
        Console.WriteLine("Вы смотрите художественный фильм");
    }

    public override void GetInfo()
    {
        Console.WriteLine($"{Title}, {Director}, {ReleaseYear}, Жанр: {Genre}");
    }

    public void Rent()
    {
        Console.WriteLine($"Художественный фильм \"{Title}\" доступен для аренды.");
    }
}

class AnimatedMovie : Movie, IStreamable, IRentable
{
    public string AnimationStudio { get; set; }

    public AnimatedMovie(string title, string director, int releaseYear, string animationStudio)
        : base(title, director, releaseYear)
    {
        AnimationStudio = animationStudio;
    }

    public override void Watch()
    {
        Console.WriteLine("Вы смотрите анимационный фильм");
    }

    public override void GetInfo()
    {
        Console.WriteLine($"{Title}, {Director}, {ReleaseYear}, Анимационная студия: {AnimationStudio}");
    }

    public void Stream()
    {
        Console.WriteLine($"Анимационный фильм \"{Title}\" доступен для потоковой передачи.");
    }

    public void Rent()
    {
        Console.WriteLine($"Анимационный фильм \"{Title}\" доступен для аренды.");
    }
}

class Cinema
{
    public string Name { get; set; }
    private List<Movie> Movies { get; set; }

    public Cinema(string name)
    {
        Name = name;
        Movies = new List<Movie>();
    }

    public void AddMovie(Movie movie)
    {
        Movies.Add(movie);
    }

    public void WatchAll()
    {
        Console.WriteLine($"Кинотеатр {Name} показывает все фильмы:\n");
        foreach (var movie in Movies)
        {
            movie.Watch();
        }
    }

    public void CompareMovies(Movie movie1, Movie movie2)
    {
        if (movie1.Rating > movie2.Rating)
        {
            Console.WriteLine($"{movie1.Title} имеет более высокий рейтинг ({movie1.Rating}) по сравнению с {movie2.Title} ({movie2.Rating})");
        }
        else if (movie1.Rating < movie2.Rating)
        {
            Console.WriteLine($"{movie2.Title} имеет более высокий рейтинг ({movie2.Rating}) по сравнению с {movie1.Title} ({movie1.Rating})");
        }
        else
        {
            Console.WriteLine($"{movie1.Title} и {movie2.Title} имеют одинаковый рейтинг ({movie1.Rating})");
        }
    }

    public void ShowMoviesInfo()
    {
        Console.WriteLine($"Фильмы в кинотеатре {Name}:");
        foreach (var movie in Movies)
        {
            movie.GetInfo();
        }
    }

    public void StreamAllMovies()
    {
        Console.WriteLine("Потоковая передача доступных фильмов:");
        foreach (var movie in Movies)
        {
            if (movie is IStreamable streamableMovie)
            {
                streamableMovie.Stream();
            }
        }
    }

    public void RentAllMovies()
    {
        Console.WriteLine("Аренда доступных фильмов:");
        foreach (var movie in Movies)
        {
            if (movie is IRentable rentableMovie)
            {
                rentableMovie.Rent();
            }
        }
    }
}
```

```
interface IReportable
{
    void GenerateReport();
}

interface ITeamLeader
{
    void GetTeamMember();
}

class Employee
{
    public string Name { get; set; }
    public string Position { get; set; }
    public int Salary { get; set; }

    public Employee(string name, string position, int salary)
    {
        Name = name;
        Position = position;
        Salary = salary;
    }

    public virtual void PrintInfo()
    {
        Console.WriteLine($"Имя: {Name}, позиция: {Position}, зарплата: {Salary}");
    }
}

class Manager : Employee
{
    public string Department { get; set; }

    public Manager(string name, string position, int salary, string department) 
        : base(name, position, salary)
    {
        Department = department;
    }

    public override void PrintInfo()
    {
        Console.WriteLine($"Имя: {Name}, позиция: {Position}, зарплата: {Salary}, Департамент: {Department}");
    }
}

class ProjectManager : Manager
{
    public string projectName { get; set; }

    public ProjectManager(string name, string position, int salary, string department, string projectName) 
        : base(name, position, salary, department)
    {
        this.projectName = projectName;
    }

    public override void PrintInfo()
    {
        Console.WriteLine($"Имя: {Name}, позиция: {Position}, зарплата: {Salary}, Департамент: {Department}, название проекта: {projectName}");
    }
}

class ProjectManagerPlus : Manager, IReportable, ITeamLeader
{
    public string projectName { get; set; }

    public ProjectManagerPlus(string name, string position, int salary, string department, string projectName) 
        : base(name, position, salary, department) 
        { 
            this.projectName = projectName;
        }

    public override void PrintInfo()
    {
         Console.WriteLine($"Имя: {Name}, позиция: {Position}, зарплата: {Salary}, Департамент: {Department}, название проекта: {projectName}");
    }
    public void GenerateReport()
    {
        Console.WriteLine("Выводит отчет");
    }

    public void GetTeamMember()
    {
        Console.WriteLine("Показывает подчиненных сотрудников");
    }
}

Employee employee = new Employee("Вася", "Уборщик", 100);
Manager manager = new Manager("Петя", "Сборщик", 200, "Министерство сборщиков");
ProjectManagerPlus projectManagerPlus = new ProjectManagerPlus("Коля", "Главный", 1000, "Министерство главных", "Проект для помощи всем людям");

employee.PrintInfo();
manager.PrintInfo();
projectManagerPlus.PrintInfo();
```
